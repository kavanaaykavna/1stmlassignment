## 4. Careful words: what I can and can't claim

**What this work can claim:**
* **Observed Correlation:** This project can demonstrate an observed statistical relationship between satellite-derived column data and ground-level air quality measurements.
* **Directional Trends:** The model can provide directional insights, indicating whether air quality is likely to improve or deteriorate over the next few hours based on incoming data.
* **Decision Support:** It can serve as a decision-support tool for city authorities to trigger health alerts or traffic restrictions when predictions cross safety thresholds.

**What this work cannot claim:**
* **Causal Proof:** This model identifies correlations; it cannot prove that one specific variable *causes* a change in AQI, as many unobserved local factors (like hyper-local traffic or construction) may be at play.
* **Absolute Accuracy:** It cannot guarantee 100% accurate predictions for every specific street corner, nor can it replace the precision of high-grade physical monitoring stations.
* **Causal Health Impact:** It cannot claim to measure the direct health outcome of an individual, only the estimated concentration of pollutants in their vicinity.
