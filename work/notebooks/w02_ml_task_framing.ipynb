## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

ML Task Type: Classification
I frame my lane as a binary classification problem because the goal is to predict whether a webpage is likely to decline in search performance or not decline based on its historical SEO metrics. Each webpage is classified into one of two categories using features such as impressions, CTR, average position, content age, and days since the last update. The predicted probability can then be used to rank pages so that the SEO team can prioritize which ones to review first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Target: The target is whether a webpage is declining or not declining in search performance. In the starter dataset, this is represented by the label is_declining_label, which is created from the trend_direction column.
Where does the label come from?
The label comes from an observed outcome, not from a prediction. It is created using a defined rule based on historical performance: if trend_direction = "down", then is_declining_label = 1 (declining); otherwise, is_declining_label = 0 (not declining). This means the model learns from what actually happened in the past and uses those examples to predict whether new webpages are likely to decline.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Success Metric: Precision@50 I will use Precision@50 as the primary success metric because the SEO team can review only a limited number of webpages at a time. Precision@50 measures the proportion of the top 50 pages recommended by the model that are actually declining. A higher Precision@50 means the model is better at identifying pages that truly need attention, helping the team use its time and resources more effectively. What number means "good"? A good model should achieve a higher Precision@50 than a simple hand-written rule or baseline. In the starter notebook, the learned model is compared against a hand rule, so improvement over that baseline demonstrates that the model is adding value. Generally, the closer Precision@50 is to 1.0 (100%), the better the model is at recommending the right pages for review.
